In the data `html.htm` Each S-star position give in the Sky-plane (geocentric equatorial coordinates) by *right ascension*- $\alpha\,\, [mas]$  and *declination*- $\delta\,\, [mas]$ and observational time given in the *fractional year*.

<img src="Ra_and_dec_on_celestial_sphere.png" width="30%" height="30%">

**Transformations** (for more details - [ref1](https://galaxiesbook.org/chapters/A.-Coordinate-systems.html), [ref2](https://arxiv.org/pdf/2007.11478.pdf) and [ref3](https://www.jameswatkins.me/posts/converting-equatorial-to-cartesian.html))

To work with physical quantities we must transform $(\alpha,\delta)$ to the cartesian $(X,Y)$ position and after that to the real position $(x,y)$:

\begin{equation*}
(\alpha,\delta)\rightarrow (X,Y)\rightarrow (x,y)
\end{equation*}


\begin{align*}
X &= D_{\odot} \cos(\delta)\cos(\alpha)\\
Y &= D_{\odot}\cos(\delta)\sin(\alpha)\\
Z &= D_{\odot}\sin(\delta)
\end{align*}

where $D_{\odot}$ is the distance between Earth and star. Then we must find real position $(x,y)$ by Cartesian position by solving system of linear algebraic equations.

\begin{align*}
X &= xB + yG \\
Y &= xA + yF
\end{align*}
where $B,A,G,F$ is the Thiele-Innes constants which connected with real orbit parameters by following:

\begin{align*}
A &= \cos(\Omega)\cos(\omega) - \sin(\Omega)\sin(\omega)\cos(i)\\
B &= \sin(\Omega)\cos(\omega) + \cos(\Omega)\sin(\omega)\cos(i)\\
F &= -\cos(\Omega)\sin(\omega) - \sin(\Omega)\cos(\omega)\cos(i)\\
G &= -\sin(\Omega)\sin(\omega) + \cos(\Omega)\cos(\omega)\cos(i)
\end{align*}

$\omega, i, \Omega$ are the argument of pericenter, the inclination between the real orbit and the observation plane and the ascending node angle respectively (this information given in `tab2.htm`).
