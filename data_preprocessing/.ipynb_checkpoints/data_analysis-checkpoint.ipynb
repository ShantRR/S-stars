```python
import sys
sys.path.append('..')

import numpy as np
import matplotlib.pyplot as plt
from os import listdir
from os.path import isfile, join
from data_utils import *
from utils import *
```

```python
T_data_path = '../Transformed_data/'
T_stars_name = [f.split('data')[1].split('.html')[0] for f in listdir(T_data_path) if isfile(join(T_data_path, f))]
```

```python
x_stars,y_stars,*_,stars_info = Stars_Info(T_stars_name)
```

## Checking different normalizations

```python
fig = plt.figure(figsize=(15,5))

for star_name in T_stars_name:
    r_star = np.sqrt(x_stars[star_name]**2 + y_stars[star_name]**2)
    phi_star = np.mod(np.arctan2(y_stars[star_name],x_stars[star_name]),2*np.pi)
    phi = np.linspace(0, 2*np.pi,200)

    p_star = stars_info.loc[star_name ]['p [Au]']
    e_star = stars_info.loc[star_name ]['e']

    x_eq,y_eq,r_eq = ellipse(phi_star,e_star,p_star)
    x_ell,y_ell,r_ell = ellipse(phi,e_star,p_star)

    x_new, y_new, e_new, p_new = EllipseNorm(x_stars[star_name], y_stars[star_name],e_star, p_star,coeff=0.2)
    r_new = np.sqrt(x_new**2+y_new**2)

    x_ell_new,y_ell_new,r_ell_new = ellipse(phi,e_new,p_new)
    
    ax1 = plt.subplot(121)
    ax2 = plt.subplot(122)

    ax1.scatter(x_stars[star_name],y_stars[star_name], label='Observed data')
    ax1.scatter(x_eq,y_eq, marker = '+', label = 'From ellipse equation')
    ax1.plot(x_ell,y_ell, label='Ellipse')

    ax2.scatter(x_new, y_new)
    ax2.plot(x_ell_new,y_ell_new, label='Ellipse')
    
    
    ax1.set_xlabel('$x$ [Au]', fontsize=18)
    ax1.set_ylabel('$y$ [Au]', fontsize=18)
    ax1.set_title('Real orbit', fontsize=18)

    ax2.set_xlabel('$\\tilde{x}$', fontsize=18)
    ax2.set_ylabel('$\\tilde{y}$', fontsize=18)
    ax2.set_title('Normalized orbits $\\tilde{S}=1$', fontsize=18)

    
#     ax1.legend()
```

```python

for star_name in T_stars_name:
    r_star = np.sqrt(x_stars[star_name]**2 + y_stars[star_name]**2)
    phi_star = np.mod(np.arctan2(y_stars[star_name],x_stars[star_name]),2*np.pi)
    phi = np.linspace(0, 2*np.pi,200)

    p_star = stars_info.loc[star_name ]['p [Au]']
    e_star = stars_info.loc[star_name ]['e']

    x_eq,y_eq,r_eq = ellipse(phi_star,e_star,p_star)
    x_ell,y_ell,r_ell = ellipse(phi,e_star,p_star)

    x_new, y_new, e_new, p_new = EllipseNorm(x_stars[star_name], y_stars[star_name],e_star, p_star,coeff=0.2)
    r_new = np.sqrt(x_new**2+y_new**2)

    x_ell_new,y_ell_new,r_ell_new = ellipse(phi,e_new,p_new)
    
    phi_diff, u_diff = DiffEqSol(e_new,p_new)
    
    fig = plt.figure(figsize=(18,5))
    
    ax1 = plt.subplot(131)
    ax2 = plt.subplot(132)
    ax3 = plt.subplot(133)

    ax1.scatter(x_stars[star_name],y_stars[star_name], label='Observed data')
    ax1.scatter(x_eq,y_eq, marker = '+', label = 'From ellipse equation')
    ax1.plot(x_ell,y_ell, label='Ellipse')

    ax2.scatter(x_new, y_new)
    ax2.plot(x_ell_new,y_ell_new, label='Ellipse')
    
    ax3.scatter(phi_star, 1/r_new, label='Observational data')
    ax3.plot(phi_diff, u_diff, label='Numerical solution of diff. eq.')
    
    ax1.set_xlabel('$x$ [Au]', fontsize=18)
    ax1.set_ylabel('$y$ [Au]', fontsize=18)
    ax1.set_title('Real orbit', fontsize=18)

    ax2.set_xlabel('$\\tilde{x}$', fontsize=18)
    ax2.set_ylabel('$\\tilde{y}$', fontsize=18)
    ax2.set_title('Normalized orbits $\\tilde{S}=1$', fontsize=18)
    
    ax3.set_xlabel('$\\varphi$', fontsize=18)
    ax3.set_ylabel('$\\tilde{u}$', fontsize=18)
    
    
    fig.suptitle('Star '+star_name, fontsize=16)
    ax1.legend()
    ax3.legend()    
    ax2.legend()    
    
    fig.savefig(star_name+'.png', dpi=300)    
```
